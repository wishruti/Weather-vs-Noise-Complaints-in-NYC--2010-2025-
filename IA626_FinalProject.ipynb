```python
# 1. Count Complaints by Temperature Range (Bin)
import csv
from collections import defaultdict

# Define temperature bins (e.g., every 5 degrees)
bins = [-10, 0, 5, 10, 15, 20, 25, 30, 35, 40]
bin_labels = [f"{bins[i]}–{bins[i+1]}°C" for i in range(len(bins)-1)]
temp_counts = defaultdict(int)

with open("311_weather_merged1.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        try:
            temp = float(row["temperature_2m"])
            for i in range(len(bins)-1):
                if bins[i] <= temp < bins[i+1]:
                    temp_counts[bin_labels[i]] += 1
                    break
        except:
            continue

# Print result
print("\n📊 Complaints by Temperature Range:")
for label in bin_labels:
    print(f"{label}: {temp_counts[label]}")
```

```python
# 2. Count Complaints by Hour of Day
from datetime import datetime

hour_counts = [0] * 24  # 0 to 23

with open("311_weather_merged1.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        try:
            created = datetime.strptime(row["Created Date"], "%m/%d/%Y %I:%M:%S %p")
            hour = created.hour
            hour_counts[hour] += 1
        except:
            continue

# Print result
print("\n🕒 Complaints by Hour of Day:")
for h in range(24):
    print(f"{h:02d}:00 → {hour_counts[h]} complaints")
```

```python
# 3. Count Complaints by Weekday
weekday_names = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
weekday_counts = [0] * 7

with open("311_weather_merged1.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        try:
            created = datetime.strptime(row["Created Date"], "%m/%d/%Y %I:%M:%S %p")
            weekday = created.weekday()  # 0 = Monday
            weekday_counts[weekday] += 1
        except:
            continue

# Print result
print("\n📅 Complaints by Weekday:")
for i in range(7):
    print(f"{weekday_names[i]}: {weekday_counts[i]}")
```

```python
# 4. Count Complaints by Precipitation Bin
precip_bins = [0, 0.1, 1.0, 2.5, 5.0, 10.0]
precip_labels = [f"{precip_bins[i]}–{precip_bins[i+1]} mm" for i in range(len(precip_bins)-1)]
precip_counts = defaultdict(int)

with open("311_weather_merged1.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        try:
            precip = float(row["precipitation"])
            for i in range(len(precip_bins)-1):
                if precip_bins[i] <= precip < precip_bins[i+1]:
                    precip_counts[precip_labels[i]] += 1
                    break
        except:
            continue

print("\n☔ Complaints by Precipitation Level:")
for label in precip_labels:
    print(f"{label}: {precip_counts[label]}")
```

```python
#1. Weekday × Hour Heatmap
import csv
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
from collections import defaultdict

# Initialize 7 days × 24 hours matrix
heatmap_data = np.zeros((7, 24))  # rows = weekdays (Mon–Sun), columns = 0–23 hours

# Read 311 dataset
with open("311_weather_merged1.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        created_str = row.get("Created Date")
        if created_str and ("AM" in created_str or "PM" in created_str):
            created = datetime.strptime(created_str, "%m/%d/%Y %I:%M:%S %p")
            weekday = created.weekday()  # Monday=0
            hour = created.hour
            heatmap_data[weekday, hour] += 1

# Plot heatmap
plt.figure(figsize=(12,6))
plt.imshow(heatmap_data, cmap='YlOrRd', aspect='auto')
plt.colorbar(label='Number of Complaints')
plt.xticks(ticks=np.arange(24), labels=[f"{h}:00" for h in range(24)], rotation=45)
plt.yticks(ticks=np.arange(7), labels=["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])
plt.title("Heatmap of Complaints by Weekday and Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Day of Week")
plt.tight_layout()
plt.savefig("figures/weekday_hour_heatmap.png")
plt.show()
```

```python
# 2. Month-wise Seasonality Plot
import csv
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
from collections import defaultdict
month_counts = [0] * 12  # Jan to Dec

# Read 311 data
with open("311_weather_merged1.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        created_str = row.get("Created Date")
        if created_str and ("AM" in created_str or "PM" in created_str):
            created = datetime.strptime(created_str, "%m/%d/%Y %I:%M:%S %p")
            month = created.month
            month_counts[month-1] += 1

# Plot
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

plt.figure(figsize=(10,6))
plt.plot(months, month_counts, marker='o')
plt.title("Noise Complaints by Month (Seasonality)")
plt.xlabel("Month")
plt.ylabel("Number of Complaints")
plt.grid(True)
plt.tight_layout()
plt.savefig("figures/seasonality_by_month.png")
plt.show()
```

```python
# 3. Weekend vs Weekday Normalized Comparison
import csv
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
from collections import defaultdict

weekday_complaints = 0
weekend_complaints = 0
weekday_hours = 0
weekend_hours = 0

# Complaints
with open("311_weather_merged1.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        created_str = row.get("Created Date")
        if created_str and ("AM" in created_str or "PM" in created_str):
            created = datetime.strptime(created_str, "%m/%d/%Y %I:%M:%S %p")
            if created.weekday() in [5,6]:  # Saturday, Sunday
                weekend_complaints += 1
            else:
                weekday_complaints += 1

# Available hours
with open("hourly_weather.csv", "r") as wf:
    reader = csv.DictReader(wf)
    for row in reader:
        dt = row.get('datetime')
        if dt:
            dt_obj = datetime.strptime(dt, "%Y-%m-%dT%H:%M")
            if dt_obj.weekday() in [5,6]:
                weekend_hours += 1
            else:
                weekday_hours += 1

# Normalize
weekday_normalized = weekday_complaints / weekday_hours
weekend_normalized = weekend_complaints / weekend_hours

# Print
print("\n📅 Normalized Weekend vs Weekday Complaints:")
print("------------------------------------------------")
print(f"Weekday: {weekday_complaints} complaints / {weekday_hours} hours = {weekday_normalized:.2f} per hour")
print(f"Weekend: {weekend_complaints} complaints / {weekend_hours} hours = {weekend_normalized:.2f} per hour")

# Plot
labels = ['Weekday', 'Weekend']
values = [weekday_normalized, weekend_normalized]

plt.figure(figsize=(6,5))
plt.bar(labels, values, color=['green', 'orange'])
plt.ylabel("Complaints per Hour")
plt.title("Normalized Noise Complaints: Weekend vs Weekday")
plt.tight_layout()
plt.savefig("figures/normalized_weekend_vs_weekday.png")
plt.show()
```

```python
# Borough Complaint Count Plot (Unnormalized)
import csv
from collections import defaultdict
import matplotlib.pyplot as plt

# Count total complaints per borough
borough_counts = defaultdict(int)

with open("311_weather_merged1.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        borough = row.get("Borough")
        if borough:
            borough_counts[borough.strip().upper()] += 1

# Sort by value (optional)
sorted_boroughs = sorted(borough_counts.items(), key=lambda x: x[1], reverse=True)

# Plot raw complaint counts
labels = [b[0] for b in sorted_boroughs]
values = [b[1] for b in sorted_boroughs]

plt.figure(figsize=(8,5))
plt.bar(labels, values, color='plum')
plt.ylabel("Total Complaints")
plt.title("Total Noise Complaints by Borough")
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig("figures/borough_total_complaints.png")
plt.show()
```
